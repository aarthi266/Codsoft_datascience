```python
X=df.drop('Class',axis=1)
X
```

```python
y=df['Class']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from imblearn.over_sampling import SMOTE
```

```python
scaler=StandardScaler()
X[['Time','Amount']]=scaler.fit_transform(X[['Time','Amount']])
sm=SMOTE(random_state=42)
X_resampled,y_resampled=sm.fit_resample(X,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# AUC-ROC Score
roc_score = roc_auc_score(y_test, y_pred)
print(f"\nAUC-ROC Score: {roc_score:.4f}")
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd

compare_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

plt.figure(figsize=(8, 5))
sns.countplot(x='Actual', data=compare_df, label='Actual', color='blue', alpha=0.6)
sns.countplot(x='Predicted', data=compare_df, label='Predicted', color='red', alpha=0.4)
plt.title('Actual vs Predicted Fraud Count')
plt.legend()
plt.show()
```
